```python
import pandas as pd
```

```python
dataset=pd.read_csv("Kidney_Disease.CSV")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.drop("age",axis=1)
dataset
```

```python
dataset.columns
```

```python
columns=['bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr','bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad','appet', 'pe', 'ane',]
dataset[columns].fillna(dataset[columns].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
def quanqual(dataset):
        quan=[]
        qual=[]
        for ColumnName in dataset.columns:

            if (dataset[ColumnName].dtype=='O'):
                qual.append(ColumnName)
            else:
                 quan.append(ColumnName)
        return quan,qual
```

```python
quan,qual=quanqual(dataset)
```

```python
quan
```

```python
qual
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
imp
```

```python
qual
```

```python
#qual= ['rbc','pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad', 'appet','pe', 'ane', 'classification']
```

```python
for col in qual:
    if col in dataset.columns:
        # Compute the mode (most frequent value)
        mode_value = dataset[col].mode()[0]
        
        # Fill missing values with the mode
        dataset[col].fillna(mode_value, inplace=True)
```

```python
cate=dataset[qual]
cate
two=[df,cate]
preprocessed=pd.concat(two,axis=1)
```
